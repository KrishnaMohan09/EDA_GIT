```python
visa_df=pd.read_csv("Visadataset.csv")
unique_labels=sorted(visa_df['case_status'].unique())
list1=[i for i in range(len(unique_labels))]
dict1=dict(zip(unique_labels,list1))
dict1
visa_df['case_status_new']=visa_df['case_status'].map(dict1)
visa_df
```

```python
visa_df=pd.read_csv("Visadataset.csv")
for j in cat_columns:
    unique_labels=sorted(visa_df[j].unique())
    list1=[i for i in range(len(unique_labels))]
    dict1=dict(zip(unique_labels,list1))
    dict1
    visa_df[f'{j}_new']=visa_df[j].map(dict1)
visa_df
```

```python
visa_df=pd.read_csv("Visadataset.csv")
for i in cat_columns:
    unique_labels=sorted(visa_df[i].unique())
    list1=[i for i in range(len(unique_labels))]
    dict1=dict(zip(unique_labels,list1))
    dict1
    visa_df[i]=visa_df[i].map(dict1)
visa_df
```

**fit and transform**

- fit means develop the logic

- transform means process the logic

- if we observe in map method first we developed dictionary logic

- then we apply that dictionary to column

- Developing the dictionary is kind of fit

- apply the dictionary to column means we are transforming the data cat to num

- whenever we are transforming or changing the data then use **fit transform**

- if we are not changing the data then only **fit**

**np.where**

- using np.where also we can change the categorical data to numerical

- But np.where is only for binary condition

- np.where is same as if-else

- if condition is True: will apply True value

- if condition is False: will apply False value

- so np.where use for only binary labels

- A column should have only two unique labels

```python
#value=='certified' replace with 0
#otherwise 1
#np.where(con,0,1)
visa_df=pd.read_csv("Visadataset.csv")
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

**One hot encoder**

- One hot encoder if one will be ON, another will be OFF

- ON represents with 1

- OFF represents with 0

- For example case status has two unique labels

  - Certified

  - Denied

- One hot encoder creates new columns which is equal to number of unique labels

- For example: case status will create two new extra columns

  - case_status_certified

  - case_status_denied

|case_status|case_status_certified|case_status_denied|
|-|-|-|
|Denied|0|1|
|Certified|1|0|
|Denied|0|1|

**Advantages**

- using one hot encoder we will get new columns

- this new column values at a time only one will be **True**, another will be **False**

- which means new columns are **independent each other**

- Independent means **no relation**

- Means no **Multi collinearity**

- Multicollinearity is an important properity in Machine learning

- **Machine learning models expect the input columns should be independent each other**

- So one hot encoder gives the no multi collinearity

**Disadvantages**

- the disadvantage is it will increase many columns

- in the above example the total columns becomes 25510

- this is happened because of case id

- in real time we will drop the id columns

- eventhough if we drop still we are getting 30 columns

- More columns means, more time to process

- we required more infrastructure (CPU-GPU-TPU)

- This is called as **curse of dimensionality**
