**power transformer**

```python
# sklearn
# preprocessing

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
transform_wage=pt.fit_transform(visa_df['prevailing_wage'].values.reshape(-1,1))
```

```python
# Assignment
# difference between yeo johnson and boxcox methods and what is yeo johnson and what is boxcox and how it works
```

- Both the Yeo-Johnson and Box-Cox methods are used for transforming data to approximate a normal distribution, 
  - which can be particularly helpful in statistical modeling and machine learning. 

**Box-cox**

- The Box-Cox transformation is a family of power transformations parameterized by a lambda () value. 
  - It is used to stabilize variance, make the data more normal distribution-like, and improve the validity of statistical tests and models.

**Yeo-Johnson**

- The Yeo-Johnson transformation is an extension of the Box-Cox transformation that can handle both positive and negative data, 
  - making it more flexible in various situations.


- Box-Cox requires input data to be strictly positive, while Yeo-Johnson supports both positive or negative data.

- By default, zero-mean, unit-variance normalization is applied to the transformed data.
