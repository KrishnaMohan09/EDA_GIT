- If you want to drop a specific column then we will use drop method

- We already know about differences between keywords and methods

- Keywords will use directly 

- Methods will come corresponding dataframe

- **Method name: drop**

df1.drop()

- Column: Which column you want to drop

- Axis: Represents the specified value is a  row or column

    - axis = 0 means rows

    - axis = 1 means column
 
    - by default axis = 0

- inplace: will provide overwrite  the dataframe or not

    - inplace = True, this indicates overwrite the dataframe

    - inplace = False,this indicates the modified dataframe will be use when you provide new variable for the operation

```python
df1.drop('x')
# we are not providing axis
# it assuming the provide value is a row
# by default axis =0 =====rows
```

```python
df1
```

```python
df1.drop('values',axis = 1)
# after this in a new cell type df1
# you will see original dataframe
```

```python
df1.drop('values',axis = 1,inplace = True)
df1
```

```python
df1.drop(4,axis = 0,inplace = True)
df1
```

```python
df1.drop('squares',axis = 1,inplace = False)
df1
```
